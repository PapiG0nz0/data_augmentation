```python
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
import numpy as np
import cv2 as cv
import os
```

```python
img = Image.open('E:\Data Augmentation\Entrenamiento\A\A_0.jpg')
plt.imshow(img)
```

Random - brightness

```python
seed1=tf.random.Generator.from_non_deterministic_state(1).normal([])
seed2=tf.random.Generator.from_non_deterministic_state(1).normal([])
seed = [seed1,seed2]
print(seed)
```

```python
img_r_brightness = tf.image.stateless_random_brightness(img, max_delta=.4, seed=seed)
plt.imshow(img_r_brightness)
```

```python
img_arr = np.array(img_r_brightness)
save_img = Image.fromarray(img_arr)
type(img_arr)
save_img.save('A0_brightness.jpg')
```

Random - Contrast

```python
img_r_contrast = tf.image.stateless_random_contrast(img, lower = 0.2 , upper = 1, seed=seed)
plt.imshow(img_r_contrast)
```

Flip left right

```python
img_flip_left_right = tf.image.flip_left_right(img)
plt.imshow(img_flip_left_right)
```

Flip up down

```python
img_flip_up_down = tf.image.flip_up_down(img)
plt.imshow(img_flip_up_down)
```

Random - HUE

```python
img_r_hue = tf.image.stateless_random_hue(img, max_delta=0.4, seed=seed)
plt.imshow(img_r_hue)
```

Random - JPEG quality

```python
img_r_jpeg_quality = tf.image.stateless_random_jpeg_quality(img, 10, 90, seed=seed)
plt.imshow(img_r_jpeg_quality)
```

Random - saturation

```python
img_r_saturation = tf.image.stateless_random_saturation(img, lower= 0.1, upper=1, seed=seed)
plt.imshow(img_r_saturation)
```
